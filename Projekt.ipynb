```python
from torchvision import models, transforms
import torch
from PIL import Image
import os

# Ładowanie modelu AlexNet
alexnet = models.alexnet(pretrained=True)
alexnet.eval()

# Definicja transformacji dla obrazu
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

# Ścieżka do folderu ze zdjęciami
folder_path = "./tmp_png/"

# Ładowanie listy plików w folderze
image_files = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f))]

# Przechodzenie przez każdy plik i dokonywanie klasyfikacji
for image_file in image_files:
    # Ładowanie obrazu
    img_path = os.path.join(folder_path, image_file)
    img = Image.open(img_path)

    # Transformacja obrazu
    img_t = transform(img)
    batch_t = torch.unsqueeze(img_t, 0)

    # Klasyfikacja przy użyciu modelu AlexNet
    with torch.no_grad():
        out = alexnet(batch_t)

    # Wczytanie etykiet klas z pliku 'imagenet_classes.txt'
    with open('imagenet_classes.txt') as f:
        classes = [line.strip() for line in f.readlines()]

    # Sortowanie wyników i wyświetlanie top 5 klas
    _, indices = torch.sort(out, descending=True)
    top5_classes = [(classes[idx], out[0, idx].item()) for idx in indices[0][:5]]

    # Wyświetlenie wyników dla danego obrazu
    print(f"Results for image {img_path}:")
    for class_name, confidence in top5_classes:
        print(f"{class_name}: {confidence:.4f}")

    print("-" * 30)
```
